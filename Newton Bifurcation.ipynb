```python
from bifurcation import *
from function_library import *
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Palatino"
})
%matplotlib inline
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1,
                               figsize=(8, 6),
                               sharex=True,
                               dpi=300)
num_points = 50000
z0 = np.random.uniform(-4, 4, num_points)
alpha = 4
beta = np.linspace(-2, 10, num_points)
draw_bifurcation(quad_newton_map,
                 quad_map_deriv,
                 1000, 300, ax1, ax2,
                 z0_vals=z0,
                 variable_arg_name="beta",
                 variable_arg_tex_name="$\\beta$",
                 alpha=4, beta=beta)

ax1.set_xlim(-1, 10)
ax1.set_ylim(-10, 10)
ax2.set_ylim(-1, 1)
fig.savefig('quadratic_bifurcation.png', format="png", bbox_inches='tight')
```
