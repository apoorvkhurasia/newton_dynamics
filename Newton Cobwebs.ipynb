```python
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Palatino"
})
%matplotlib inline

from function_library import *
from cobweb import *
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, 
                               figsize=(12, 6), 
                               sharey=True)
alpha = 4
betas = [2, 2]
z0 = [8, -8]
draw_cobweb(quad_newton_map, quad_map_discontinuity,
            z0 = 8, steps = 10, 
            ax = ax1, 
            xlim = [-10,10], 
            ylim = [-10, 10],
            alpha = alpha, beta = betas[0])
draw_cobweb(quad_newton_map, quad_map_discontinuity,
            z0 = -8, steps = 100, 
            ax = ax2, 
            xlim = [-10,10], 
            ylim = [-10, 10],
            alpha = alpha, beta = betas[1])

ax1.set_title(f"$z_0 = {z0[0]}, \\alpha = {alpha}, \\beta = {betas[0]}$", 
              fontsize=18)
ax2.set_title(f"$z_0 = {z0[1]}, \\alpha = {alpha}, \\beta = {betas[1]}$", 
              fontsize=18)
ax2.set_ylabel('')
fig.savefig('cobweb_quad_converge.svg', format = "svg", bbox_inches='tight')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, 
                                    figsize=(12, 6), 
                                    sharey=True)
alphas = [2, 2, 2]
z0 = [3, -3, 0.5]
draw_cobweb(cubic_newton_map, cubic_map_discontinuity,
            z0 = z0[0], steps = 25, 
            ax = ax1, 
            xlim = [-4, 4], 
            ylim = [-3, 3],
            alpha = alphas[0])
draw_cobweb(cubic_newton_map, cubic_map_discontinuity,
            z0 = z0[1], steps = 25, 
            ax = ax2, 
            xlim = [-4, 4], 
            ylim = [-3, 3],
            alpha = alphas[1])
draw_cobweb(cubic_newton_map, cubic_map_discontinuity,
            z0 = z0[2], steps = 25, 
            ax = ax3, 
            xlim = [-4, 4], 
            ylim = [-3, 3],
            alpha = alphas[2])
ax1.set_title(f"$z_0 = {z0[0]}, \\alpha = {alphas[0]}$", fontsize=18)
ax2.set_title(f"$z_0 = {z0[1]}, \\alpha = {alphas[1]}$", fontsize=18)
ax3.set_title(f"$z_0 = {z0[2]}, \\alpha = {alphas[2]}$", fontsize=18)
ax2.set_ylabel('')
ax3.set_ylabel('')
fig.savefig('cobweb_cubic_converge.svg', format = "svg", bbox_inches='tight')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, 
                                    figsize=(12, 6),
                                    sharey=True)
alpha = 2
draw_cobweb(cubic_newton_map, cubic_map_discontinuity,
            z0 = -np.sqrt(alpha/5), steps = 10, 
            ax = ax1, 
            xlim = [-4, 4], 
            ylim = [-3, 3],
            alpha = alpha)
draw_cobweb(cubic_newton_map, cubic_map_discontinuity,
            z0 = -np.sqrt(alpha/5) + 0.001, steps = 10, 
            ax = ax2, 
            xlim = [-4, 4], 
            ylim = [-3, 3],
            alpha = alpha)
draw_cobweb(cubic_newton_map, cubic_map_discontinuity,
            z0 = -np.sqrt(alpha/5) - 0.001, steps = 10, 
            ax = ax3, 
            xlim = [-4, 4], 
            ylim = [-3, 3],
            alpha = alpha)

ax1.set_title("$z_0 = -\\sqrt{\\alpha/5}, \\alpha = 2$", fontsize=18)
ax2.set_title("$z_0 = -\\sqrt{\\alpha/5} + \\epsilon, \\alpha = 2$", fontsize=18)
ax3.set_title("$z_0 = -\\sqrt{\\alpha/5} - \\epsilon, \\alpha = 2$", fontsize=18)
ax2.set_ylabel('')
ax3.set_ylabel('')
fig.savefig('cobweb_cubic_cycle_sensitive_neg.svg', format = "svg", bbox_inches='tight')
```
